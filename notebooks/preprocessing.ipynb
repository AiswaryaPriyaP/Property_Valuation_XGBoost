```python
import os
import pandas as pd
import numpy as np

BASE_DIR = r"C:\Users\Aiswarya\Desktop\Property_Valuation_XGBoost"
DATA_DIR = os.path.join(BASE_DIR, "data")
```

```python
train_df = pd.read_excel(os.path.join(DATA_DIR, "train(1).xlsx"))
test_df = pd.read_excel(os.path.join(DATA_DIR, "test2.xlsx"))

train_df.shape, test_df.shape
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(train_df["price"], bins=50)
plt.xlabel("Property Price")
plt.ylabel("Number of Properties")
plt.title("Distribution of Property Prices")
plt.tight_layout()
plt.show()
```

```python
TARGET = "price"

FEATURES = [
    "bedrooms",
    "bathrooms",
    "sqft_living",
    "sqft_lot",
    "floors",
    "condition",
    "grade",
    "lat",
    "long",
    "view",
    "waterfront"
]

X = train_df[FEATURES]
y = train_df[TARGET]

X_test = test_df[FEATURES]

X.shape, y.shape, X_test.shape
```

```python
X.to_csv(os.path.join(DATA_DIR, "train_features.csv"), index=False)
X_test.to_csv(os.path.join(DATA_DIR, "test_features.csv"), index=False)
y.to_csv(os.path.join(DATA_DIR, "train_target.csv"), index=False)

print("Preprocessing complete for XGBoost pipeline.")
```
