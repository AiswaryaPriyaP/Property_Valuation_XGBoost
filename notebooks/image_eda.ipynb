```python
## Satellite Image EDA

This notebook presents a random sample of satellite image tiles fetched using
property latitude and longitude coordinates. The objective is to visually inspect
the quality and variability of images used for qualitative analysis and reporting.
```

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python
IMG_DIR = r"C:\Users\Aiswarya\Desktop\Property_Valuation_XGBoost\images"
```

```python
image_files = os.listdir(IMG_DIR)

sample_imgs = random.sample(image_files, min(9, len(image_files)))

plt.figure(figsize=(6,6))
for i, img_name in enumerate(sample_imgs):
    img_path = os.path.join(IMG_DIR, img_name)
    img = Image.open(img_path)
    plt.subplot(3,3,i+1)
    plt.imshow(img)
    plt.axis("off")

plt.suptitle("Sample Satellite Images")
plt.show()
```
